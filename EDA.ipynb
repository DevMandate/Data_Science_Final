```python
# Exploratory Data Analysis
# We shall use pandas to read our CSV file (bank)

import pandas

data = pandas.read_csv('datasets/bank.csv')
data
```

```python
# printing ot a single column 
data['Gender']
```

```python
# check whether you have empty sets
data.isnull().sum()
```

```python
# Filling the empty spots on the self_employed column
data['Self_Employed'].fillna('unknown', inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head(50)
```

```python
mean_loan = data['LoanAmount'].mean()
mean_loan
```

```python
# Replace all the empty slots in the loan_mean column with the mean gotten
data['LoanAmount'].fillna(mean_loan, inplace=True)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt 
```

```python
# Checking versions available on matplotib library
plt.style.available
```

```python
# use one of the versions
plt.style.use(['seaborn-v0_8'])
```

```python
# by use of matplotlib, we draw a count plot bar
# A count plot is a univariate meaning it only accepts one column
sn.countplot(x= data['Property_Area'], palette='spring')
plt.title('Property Area Owned by HHDS')
plt.ylabel('Total Number of HHDS')
plt.xlabel('HHDS Setting')
```

```python
# Below we are going to import Numpy linrary that will enable us to work with arrays and matrices
import numpy
```

```python
# creating bar plots
# Bar plots are  bivariate or multivariate - they accept two or more columns
# Note:  A barplot is drawn by use of a categorical data combined with a quantitatve data/column

sn.barplot(x='Property_Area', y='LoanAmount', data=data, estimator=numpy.mean, palette='winter', hue='Married')
```

```python
# A swarm plot
# A swarm plot is a plot that helps to visualize where more members of a certain group lies/fall
# It shows the distribution of different members
# is a bivariate or multivariate 

sn.swarmplot(x='Gender', y='ApplicantIncome', data=data, palette='magma', hue='Education')
```

```python
# Scatter plot (Quantitative columns only)
# It is used to show distribution of different entities of a column
# It is also used to show the relationship between the 2 variables

sn.scatterplot(x='LoanAmount', y='ApplicantIncome', data=data, palette='spring', hue='Property_Area')
```

```python
# PieChart
# It is univariate

data.groupby('Loan_Status').size().plot(kind="pie", autopct = '%1.1f%%')
```

```python
data.groupby('Loan_Status').size().plot(kind="line")
```
