```python
#import pandas and read the data
import pandas
data = pandas.read_csv('datasets/Mall_Customers.csv')
data
```

```python
# check for empty slots
data.isnull().sum()
```

```python
# Drop/ignore the empty slots on the data
data.dropna(inplace=True)
```

```python
# check for empty slots again
data.isnull().sum()
```

```python
# Create an array where you replace your predicts
array = data.values
our_data = array[:, 2:5]
our_data
```

```python
# when dealing with numerical/quantitave data, you use KMeans
# For categorical data, we use KModes
# For the combination, we use KPrototype
from sklearn.cluster import KMeans
model = KMeans(10, random_state=42)
model.fit(our_data)
```

```python
# List out the groups / clusters created by the model
means = model.cluster_centers_
means
```

```python
# Use a dataframe to represent the clusters
clusters = pandas.DataFrame(means, columns=['Age', 'Income', 'Spending_Score'])
clusters
```

```python
# Show members of one of the clusters
data['Cluster_number'] = model.labels_
clusters = data[data['Cluster_number'] == 3]
clusters
```

```python
# Export members of a particular clusterinto a csv
data[data['Cluster_number'] == 3].to_csv('Cluster_3_member.csv')
```
