```python
# Step 1: Convert the 'Spending_Score' column to numeric, coercing errors to NaN
data['age'] = pandas.to_numeric(data['age'], errors='coerce')
data['annual_income'] = pandas.to_numeric(data['annual_income'], errors='coerce')
data['spending_score'] = pandas.to_numeric(data['spending_score'], errors='coerce')
```

```python
# Step 2: Calculate the mean of the columns
mean_value = data['age'].mean()
mean_value = data['annual_income'].mean()
mean_value = data['spending_score'].mean()
```

```python
# Step 3: Replace NaN values with the mean of the column
data['age'].fillna(mean_value, inplace=True)
data['annual_income'].fillna(mean_value, inplace=True)
data['spending_score'].fillna(mean_value, inplace=True)
```

```python
# Calculate correlation
correlation = data['age'].corr(data['annual_income'])
print(f'Correlation between age and annual income: {correlation}')
```

```python
import statsmodels.api as sm

# Prepare the data for regression
X = data['age']  # Independent variable
y = data['annual_income']  # Dependent variable

# Add a constant to the independent variable
X = sm.add_constant(X)

# Create and fit the model
model = sm.OLS(y, X).fit()

# Print the regression results
print(model.summary())
```
